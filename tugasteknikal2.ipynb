# Import library sqlite3 and pandas

```python
import sqlite3
import pandas as pd
```

# Connect sqlite3 to python

```python
connection = sqlite3.connect('data.sqlite')
result = connection.cursor()
```

# Activity

## Tampilkan data customer yg mempunyai opening amount diatas 7000.

```python
command = "SELECT * FROM customer WHERE OPENING_AMT > '7000.00';"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Tampilkan informasi order yg dilakukan melalui agensi dengan code A003.

```python
command = "SELECT * FROM orders WHERE AGENT_CODE = 'A003';"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Tampilkan data customer yg tinggal di New York, Bangalore, Mumbai, dan Torento.

```python
command = "SELECT * FROM customer WHERE CUST_CITY IN ('New York', 'Bangalore', 'Mumbai', 'Torento');"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Tampilkan informasi customer yang mempunyai payment amount diatas 7000 dan urutkan berdasarkan opening amountnya dari yg terbesar

```python
command = "SELECT * FROM customer WHERE PAYMENT_AMT > '7000.00' ORDER BY OPENING_AMT DESC;"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Tampilkan data agensi yang tidak bekerja di london dan memiliki comission lebih dari 0.12

```python
command = "SELECT * FROM agents WHERE WORKING_AREA NOT IN ('London') AND COMMISSION > '0.12';"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```
