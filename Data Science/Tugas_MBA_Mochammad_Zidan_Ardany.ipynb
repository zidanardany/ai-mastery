# Market Basket Analysis

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
transaction_list = []

for i in range(data.shape[0]):
  tlist = [j for j in data.iloc[i] if str(j) != 'nan']
  transaction_list.append(tlist)

print(len(transaction_list),'\n')

transaction_list[:5]
```

```python
te = TransactionEncoder()
te_ary = te.fit(transaction_list).transform(transaction_list)
data2 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(data2, min_support=0.03, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold= 1.0)
metrix = pd.DataFrame(rules)
metrix[['antecedents','consequents', 'support', 'confidence', 'lift']].sort_values('lift', ascending=False)
```

```python
metrix['consequents'].value_counts().head()
```

```python
metrix['antecedents'].value_counts().head()
```

# Kesimpulan

Dari analisis diatas terlihat bahwa semakin tinggi nilai peningkatan, semakin kuat korelasi antar item. Data menunjukkan bahwa mineral water adalah anteseden dan konsekuensi populer, yang masuk akal karena itu adalah barang yang paling banyak terjual.

Selain mineral water, kita dapat menyarankan beberapa paket bundling untuk mengoptimalkan penjualan barang tersebut berdasarkan korelasi item tertinggi (format: antecedents -> consequent, dengan urutan: lift, kemudian confidence tertinggi):

> Spaghetti --> Ground Beef

> Milk --> Spaghetti

> Milk --> Chocolate

> Chocolate --> Spaghetti

> Chocolate --> French Fries

Selain paket bundling, bisnis dapat membuat keputusan berdasarkan temuan ini. Misalnya, Market ingin meletakkan Spaghetti atau French Fries di dekat Chocolate dan Milk karena orang yang membeli susu (Milk) tampaknya juga tertarik membeli Chocolate, Spaghetti dan French Fries
