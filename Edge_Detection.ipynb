```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('dave.jpg', 0)

plt.imshow(img, 'gray')
```

cpu : bagus untuk perhitungan aritmatik, integral, differential, dll

gpu : cepat untuk matriks

```python
# Create custom kernel

sobel_y = np.array([
                    [-1,-2,-1],
                    [0, 0, 0],
                    [1, 2, 1]
                  ])
filtered_img_y = cv2.filter2D(img, -1, sobel_y)
plt.imshow(filtered_img_y, 'gray')
```

```python
# Create custom kernel

sobel_x = np.array([
                    [-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]
                  ])
filtered_img_x = cv2.filter2D(img, -1, sobel_x)
plt.imshow(filtered_img_x, 'gray')
```

```python
# Create custom kernel

sobel_x = np.array([
                    [0, 1, 0],
                    [1, -4, 1],
                    [0, 1, 0]
                  ])
filtered_img_x = cv2.filter2D(img, -1, sobel_x)
plt.imshow(filtered_img_x, 'gray')
```

```python
'''https://docs.opencv.org/4.x/d5/d0f/tutorial_py_gradients.html'''

laplacian = cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(laplacian, 'gray')
```
