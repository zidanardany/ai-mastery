# 4Ws of Problem Scoping

## WHO
Siapa yang memiliki masalah tersebut?

1. Siapa saja yang mengalami masalah tersebut?

  --> Pasien dewasa mulai dari umur 30 terutama lansia

2. Siapa saja yang terlibat dalam masalah tersebut?

  --> Pasien dewasa, dokter

3. Apa yang kamu ketahui tentang mereka?
  
  --> mereka memiliki resiko penyakit coronary artery disease (CAD)

## WHAT
Apa permasalahan sebenarnya?

1. Apa masalah utamanya?

  --> Penduduk usia dewasa pada umumnya tidak dapat mengidentifikasi apakah mereka memiliki penyakit jantung (CAD) secara mandiri. Identifikasi hanya bisa dilakukan oleh dokter menggunakan alat kesehatan khusus. Akan tetapi, jumlah dokter spesialis jantung dan alat kesehatan jantung terbatas dan biaya untuk tes kesehatan jantung lumayan mahal. Padahal penduduk usia dewasa berpotensi terkena serangan jantung. Hal ini dapat menyebabkan terlambatnya penanganan kasus.

2. Apa faktor-faktor pendukung masalah tersebut?

  --> "Di 2014, dari 52 persen angka kematian di seluruh dunia akibat penyakit tidak menular ternyata yang terbanyak disebabkan penyakit jantung" (CNN Indonesia).

  --> Menurut data Riskesdas Indonesia tahun 2018, 4,2 juta orang menderita penyakit kardiovaskular

  --> IHME melaporkan 14,4% penyebab kematian di Indonesia adalah penyakit CAD.

  faktor yang memperlambat diagnosa CAD:
  *   jumlah dokter spesialis jantung dan alat kesehatan jantung terbatas
  *   biaya untuk tes kesehatan jantung lumayan mahal

  faktor yang dapat menyebabkan seseorang beresiko terkena CAD (SehatQ):
  *   Menghisap rokok
  *   Jarang bergerak
  *   Memiliki stress tinggi

Sumber:
*   http://labdata.litbang.kemkes.go.id/images/download/laporan/RKD/2018/Laporan_Nasional_RKD2018_FINAL.pdf
*   https://www.sehatq.com/artikel/cad-adalah-salah-satu-penyakit-jantung-yang-mematikan
*   https://www.cnnindonesia.com/gaya-hidup/20210929192536-255-701239/angka-kematian-akibat-jantung-masih-lebih-tinggi-dari-covid


## WHERE
Dimana/pada saat apa permasalahan ini muncul?

1. Dimana saja si calon "user" mengalami masalah ini?

  --> Umumnya di daerah perkotaan, karena biasanya penduduk kota jarang bergerak dan memiliki stress tinggi

2. Pada kondisi seperti apa si calon "user" mengalami masalah ini?

  --> Saat seseorang berusia 50 tahun ke-atas dan memiliki ciri-ciri terkena CAD namun ia mengalami keterbatasan dalam menjangkau fasilitas kesehatan

## WHY
Mengapa kamu yakin masalah ini sangat penting dibahas?

1. Apa nilai utama permasalahan ini?

  --> Dapat membantu pasien yang memiliki keterbatasan dalam menjangkau fasilitas kesehatan untuk mengidentifikasi permasalahan mereka secara akurat.

  --> Dapat membantu pasien untuk mengidentifikasi CAD lebih dini, sehingga mereka mendapatkan penanganan lebih awal

2. Bagaimana kamu bisa memperbaiki (sebagian atau seluruh) dari masalah tersebut?

```python
print('Ada Boost accuracy:', acc_score_abc)
print('Ada Boost precission:', prec_score_abc)
```

Terlihat bahwa akurasi model Ada Boost pada data yang belum dinormalisasi adalah 78% dan presisi 56%

## Model Pengembangan dengan Stratified K-Fold
