```python
valid = ['batu', 'gunting', 'kertas']

pemain1 = input('pilihan pemain 1: ')
pemain2 = input('pilihan pemain 2: ')

if pemain1 in valid and pemain2 in valid:
  if pemain1 != pemain2:
    if pemain1 == 'batu':
      if pemain2 == 'gunting':
        print('pemain 1 menang!')
      else:
        print('pemain 2 menang!')
    elif pemain1 == 'gunting':
      if pemain2 == 'kertas':
        print('pemain 1 menang!')
      else:
        print('pemain 2 menang!')
    elif pemain1 == 'kertas':
      if pemain2 == 'batu':
        print('pemain 1 menang!')
      else:
        print('pemain 2 menang!')
  else:
    print('hasil seri!')
else:
  print('input tidak valid')
```

## No 2

```python
menu = {'1': ['Donat', '1250'], '2': ['Bakwan', '1000'], '3': ['Onde-onde', '800'], '4': ['Lemper', '1250'], '5': ['Risol', '1500']}

print('''MENU JAJANAN PASAR
1. Donat - Rp 1250
2. Bakwan - Rp 1000
3. Onde-onde - Rp 800
4. Lemper  - Rp 1250
5. Risol - Rp 1500''')

while True:
  pesanan = str(input('pesan apa maszeh? '))

  if pesanan == 'cukup':
    break
  elif pesanan in menu.keys():
    print(f"kamu memesan {menu[pesanan][0]} dengan harga Rp {menu[pesanan][1]}, terima kasih sudah berbelanja")
  else:
    print('Tidak ada di daftar menu')

print('Terima kasih sudah berbelanja jajanan pasar kami!')
```

## No 3

```python
menu = {'1': ['Donat', 1250], '2': ['Bakwan', 1000], '3': ['Onde-onde', 800], '4': ['Lemper', 1250], '5': ['Risol', 1500]}

print('''MENU JAJANAN PASAR
1. Donat - Rp 1250
2. Bakwan - Rp 1000
3. Onde-onde - Rp 800
4. Lemper  - Rp 1250
5. Risol - Rp 1500''')

while True:
  pesanan = str(input('pesan apa maszeh? '))

  if pesanan == 'cukup':
    break
  elif pesanan in menu.keys():
    jumlah = int(input('berapa buah? '))
    print(f"kamu memesan {menu[pesanan][0]} dengan jumlah {jumlah} dengan harga Rp {menu[pesanan][1] * jumlah}, terima kasih sudah berbelanja")
  else:
    print('Tidak ada di daftar menu')

print('Terima kasih sudah berbelanja jajanan pasar kami!')
```
