#### sepal_length

```python
# Atur gambar matplotlib
f, axes = plt.subplots(ncols=2, figsize=(10,5), sharex=True)
sns.despine(left=True)

# Plot Swarmplot
sns.swarmplot(x=iris['class'], y=iris['sepal_length'], ax=axes[0])

# Plot Boxplot
sns.boxplot(x=iris['class'], y=iris['sepal_length'], ax=axes[1])

plt.tight_layout()
```

Berdasarkan kedua plot di atas, *sepal length* tampaknya merupakan fitur yang cukup menarik untuk mengklasifikasikan spesies yang berbeda. Kita dapat mengetahui bahwa:

*   Meskipun perbedaan *sepal length* antara Iris-versicolor dan Iris-virginica kurang jelas, sebagian besar *sepal length* bunga Iris-versicolor lebih kecil daripada *sepal length* bunga Iris-virginica.

## Pengenalan Regresi Linier
Untuk memilih garis terbaik, kita dapat mendefinisikan fungsi yang mengukur seberapa dekat atau jauh prediksi (nilai pada garis) dari nilai sebenarnya, ini dikenal sebagai **cost function**. 

Idenya adalah untuk meminimalkan cost function untuk mendapatkan garis terbaik. Untuk meminimalkannya, kita dapat menggunakan algoritma gradient descent. 

Inilah garis besar langkah yang dilakukan oleh algoritma regresi linier.

### Regresi Linier di sklearn

```python
X = data.LotArea[:,np.newaxis] # Reshape
y = data.SalePrice

# Fitting Regresi Linier Sederhana pada data latih
regressor = LinearRegression()
regressor.fit(X, y)

LinearRegression()
```

```python
# Memprediksi data uji
y_pred = regressor.predict(X)
```

```python
# Plotting Garis Fitting Terbaik
plt.scatter(X, y)
plt.plot(X, y_pred, 'r-')
plt.title('Housing Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

### Buat Prediksi

Misalkan kita ingin membeli rumah seluas 14.000 kaki persegi, maka kita cukup menggambar garis vertikal dari 14.000 hingga garis Tren Perkiraan kita dan melanjutkan garis itu menuju sumbu y

```python
plt.scatter(X, y)
plt.plot(X, y_pred, 'r-')
plt.title('Housing Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.axvline(x=14000, c='g')
plt.show()
```

Terlihat bahwa untuk rumah yang luasnya ~ 14.000 kita perlu membayar ~ 200.000-225.000 dolar

## Regresi Linier Multivariat

Akan digunakan lebih dari satu variabel Independen dan akan memprediksi variabel Dependen yaitu Harga

```python
!wget 'https://raw.githubusercontent.com/ksnugroho/ai-mastery/main/dataset/house_prices_multivariate.csv'
```

```python
# Load Data

NY_Housing = pd.read_csv('house_prices_multivariate.csv')
NY_Housing.head()
```

```python
# Membuat variabel independent dan dependent dari dataset
X = NY_Housing.iloc[:,:-1] # Pilih semua kolom kecuali kolom terakhir
y = NY_Housing.SalePrice

# Fitting Regresi Linier Berganda
regressor = LinearRegression()
regressor.fit(X, y)

LinearRegression()

print('intercept', regressor.intercept_) # y-intercept
print('koefisien predictor', regressor.coef_) # bobot/koefisien regresi
```

## Memprediksi harga

Percobaan memprediksi harga sebuah rumah

```python
my_house = X.iloc[155]

my_house
```

```python
pred_my_house = regressor.predict(my_house.values.reshape(1, -1))

print('harga yang diprediksi', pred_my_house[0])
```

```python
print('harga sebenarnya', y[155])
```

Terlihat bahwa harga prediksi tidak terlalu jauh dari harga sebenarnya

Prediksi harga seluruh rumah pada dataset

```python
# Memprediksi hasil
y_pred = regressor.predict(X)
y_pred[:10]
```

```python
prices = pd.DataFrame({'actual': y, 'predicted': y_pred})
prices.head(10)
```

## Mengukur tingkat kecocokan (goodness of fit)
Mengukur goodness of fit dapat dilakukan dengan cost function

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

Ada tiga metrik evaluasi dasar untuk memeriksa tingkat kecocokan regresi:
*   Mean Absolute Error
*   Root Mean square Error
*   R-Square (Nilai sisa)
